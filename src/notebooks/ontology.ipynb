```python
import json


base_url = "https://api.dowjones.com"
headers = {
    "user-key": user_key,
    "Content-Type": "application/json",
    "X-API-VERSION": "3.0"
}

def check_response(response):
    if response.status_code in [200, 201]:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        print(response.text)
        return None
    
    
def create_code(query_body):
    """
    Create a code request to check how many documents match a query
    """
    url = f"{base_url}/code/search/format"
    response = requests.get(url, headers=headers, data=json.dumps(query_body))
    return check_response(response)

```

```python
query = {"format": "json", "keywords": "ibm", "Category": "ticket"}
resp = create_code(query)
```
